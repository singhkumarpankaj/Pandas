```python
def cat_Fare(x):
    if x < 10:
        return "cheap"
    elif x >= 10 and x < 20:
        return 'mid'
    else:
        return 'high'
```

```python
df['cat_Fare']  = df['Fare'].apply(cat_Fare)
```
